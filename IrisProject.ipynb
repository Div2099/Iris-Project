```python
import pandas as pd
import seaborn as sns 
```

```python
#Dataset
df = pd.read_csv("Iris.csv")
df.columns 
df.info()
df.describe()
df
```

```python
#1 Missing values check (Cleaning)
print("Missing values per column:")
print(df.isnull().sum())
```

```python
#2 Duplicates if any
print(f"\nNumber of Duplicate rows:{df.duplicated().sum()}")
df = df.drop_duplicates()
```

```python
#3 Data types (Formatting)
print("\nData Types:")
print(df.dtypes)
```

```python
#Average per species 
a = df.groupby('Species').mean()
print("\nAverage Measurements Per Species:")
print(a)
```

```python
#Finding Trend: Correlation 
c = df.corr(numeric_only=True)
print("\nCorrelation Matrix:")
print(c)
```

```python
#Histogram 

import matplotlib.pyplot as plt 
plt.figure(figsize = (10,6))
sns.histplot(data=df, x='PetalLengthCm',hue='Species',kde=True)
plt.title('Distribution of Petal Length by Species')
plt.show()
```

```python
#Comparing Averages

plt.figure(figsize=(10,6))
sns.boxplot(x='Species', y='SepalWidthCm',data=df)
plt.title('Comparing Of Sepal Width Across Species')
plt.show()
```
